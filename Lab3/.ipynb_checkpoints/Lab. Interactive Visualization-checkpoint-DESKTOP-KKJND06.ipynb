```python
%matplotlib inline
```

## 3. Widget GUI Tools

Together with the widget magic command, we can declare and implement extra functionlity on our plots using different GUI elements with the [interact()](https://ipywidgets.readthedocs.io/en/stable/examples/Using%20Interact.html) function from ipywidgets.

The GUI elements availbel are:

### Sliders:

```python
def f(x):
    return x

interact(f, x=10);
```

### Check boxes:

```python
def f(x):
    return x

interact(f, x=True);
```

### Text box:

```python
def f(x):
    return x

interact(f, x='Insert text here');
```

### Dropdown menu:

```python
def f(x):
    return x

interact(f, x=[1,2,3,4]);
```

All these widgets can be used together by presenting the *interact()* function with multiple arguments. However, the interact function must be declared as a [python decorator](https://www.python-course.eu/python3_decorators.php):

```python
@interact(x=True, y=1.0, z=[1,2,3])
def g(x, y, z):
    return (x, y, z)
```

By including these widget GUI elements in our interactive plots, we can expand the functionality of how the data is respresented and presented to the viewer. Let's build an example implementing the *reconfiguring* method with polynomial regressions. First we will create an scatter plot using the data stored in the *gdp* and *corruption* variables:

```python
deg = 1
data_reg = np.polyfit(gdp, corruption, deg)
xp = np.linspace(0.0,1.75,156)
yp = np.polyval(data_reg, xp)

fig = plt.figure(figsize=(9.5,5))
ax = fig.add_subplot()
ax.scatter(gdp, corruption, label='GDP vs Corruption', color='gray')
ax.plot(xp, yp, label='Pol. Reg.')
ax.set_title('GDP vs Corruption')
ax.set_xlabel('GDP')
ax.set_ylabel('Corruption')
ax.grid()
ax.legend()
```

For the previous example we have used a level 1 for the polynomial regression. Now we will implement a slider that returns a number between 1 and 5 and use that number to update the level of our polynomial regression. To do so we will implement 2 functions:
1. f will be our main plotting function. It will recieve the value from the slider, calculate the polynomial regression, and plot the data.
2. erase_plot is a function to clear the canvas on the figure, before plotting data with a new polynomial regression level.

```python
fig = plt.figure(figsize=(9.5,5))
ax = fig.add_subplot()

def erase_plot():
    ax.clear()

def f(reg_lvl):
    erase_plot()
    deg = reg_lvl
    data_reg = np.polyfit(gdp, corruption, deg)
    xp = np.linspace(0.0,1.75,156)
    yp = np.polyval(data_reg, xp)
    scatter = ax.scatter(gdp, corruption, label='GDP vs Corruption', color='gray')
    ax.plot(xp, yp, label='Pol. Reg.')
    ax.grid()
    ax.legend()
    ax.set_title('GDP vs Corruption')

interact(f, reg_lvl=widgets.IntSlider(min=1, max=5, step=1, value=1));
```

We will implement another example by using checkboxes for filtering, and using colors for econding our data. We will plot the data from *gdp* against the data from *social*, *life*, and *corruption*

```python
fig = plt.figure(figsize=(9.5,5))
ax = fig.add_subplot()
ax.scatter(gdp, social, label='Social', color='orange')
ax.scatter(gdp, life, label='Life', color='blue')
ax.scatter(gdp, corruption, label='Corruption', color='red')
ax.set_xlabel('GDP')
ax.grid()
ax.legend()
ax.set_title('GDP vs Social, Life and Corruption')
```

```python
fig = plt.figure(figsize=(9.5,5))
ax = fig.add_subplot()

def erase_plot():
    ax.clear()

def f(social_scatter, life_scatter, corruption_scatter):
    erase_plot()
    if(social_scatter==True):
        ax.scatter(gdp, social, label='Social', color='orange')
        
    if(life_scatter==True):
        ax.scatter(gdp, life, label='Life', color='blue')
        
    if(corruption_scatter==True):
        ax.scatter(gdp, corruption, label='Corruption', color='red')
    else:
        ax.scatter(gdp, corruption, label='Corruption', color='gray', alpha=0.5)
        
    ax.set_xlabel('GDP')
    ax.grid()
    ax.legend()
    ax.set_title('GDP vs Social, Life and Corruption')

interact(f, social_scatter=False, life_scatter=False, corruption_scatter=False);
```

## 4. Animated Plots

In addition to the widget functionality, matplotlib offers native support for animated plot through its [animation](https://matplotlib.org/stable/api/animation_api.html) class. In the following example we will animate a sine function using matplotlib. 

First we will create the data space for the x axis in *x_val* variable, and store the values of *sin(x)* in the *y_val* variable. Then we will animate 2 different plots: one for the *dot* representing the current value of a pair (x,y), and another for the lines that will show the trayectory covered by the dot along the function (*linegraph*). Next we will define a function called *animate* that will recieve a parameter *interval* that usually control the refresh rate of the animation, but that we will use to navigate through the numerical space created in *x_val*. Inside the animate function we will update the values that should be computed for the ploting of linegraph and dot. Lastly, we store the animation data within a variable called *anim*, from where the animate function will be called.

```python
x_val = np.linspace(0, 2*np.pi, 100)
y_val = np.sin(x_val)

fig = plt.figure(figsize=(9.5,5))
ax = fig.add_subplot()
ax.set_xlim([0, 2*np.pi])
ax.set_ylim([-1.1, 1.1])

linegraph, = ax.plot([], [], color='pink')
dot, = ax.plot([], [], 'o', color='red')

def animate(interval):
    linegraph.set_data(x_val[:interval],y_val[:interval])
    dot.set_data(x_val[interval],y_val[interval])

anim = animation.FuncAnimation(fig, animate, frames=len(x_val), interval=50)
```

Once our animation data is stored within amin, we can render this information into an HTML5 video (ffmpeg), or a javascript widget from the [Animation class](https://matplotlib.org/stable/api/_as_gen/matplotlib.animation.Animation.html) in matplolib:

```python
HTML(anim.to_html5_video())
```

```python
HTML(anim.to_jshtml())
```

This final exmaple will allow you to plot data in real time by reading the information written in an external file called *animate.txt*.

You can make changes directly to the file and, after you saved it, matplolib will automatically update the plot with the new data in the file.

```python
fig = plt.figure(figsize=(9.5,5))
ax = fig.add_subplot()

def animate(interval):
    ani_data = open('animate.txt', 'r').read()
    lines = ani_data.split('\n')
    x_values = []
    y_values = []
    for line in lines:
        if len(line) > 1:
            x, y = line.split(',')
            x_values.append(int(x))
            y_values.append(int(y))
    ax.clear()
    ax.plot(x_values,y_values)
    
ani = animation.FuncAnimation(fig, animate, interval=1000)
```
